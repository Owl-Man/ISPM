# **Задание 1. Определение концептуальных основ дипломного проекта в области информационных систем**



В рамках дисциплины "Управление проектами информационных систем" представьте развернутое концептуальное обоснование вашего проекта, ответив на следующие вопросы:


#### **1. Обоснование необходимости разработки**


Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?




*Обоснуйте актуальность решаемой проблемы для конкретных заинтересованных сторон.*

<font color='red'><b>Ваш ответ</b></font>

Для игроков и разработчиков игр. Призвана удовлетворить потребности игроков, а также потребности разработчиков игр в создании ИИ-оппонентов ("ботов"). Данное решение имеет значимую актуальность, поскольку задача создания ИИ-оппонентов при разработке игр встречается часто. В особенности без этого не обойтись в стратегиях, ведь даже если в игре реализована онлайн составляющая, где противниками являются настоящие игроки, обычной практикой является реализация также одиночной игры против ботов. В связи с чем создание решений для данной задачи является актуальным для разработчиков игр.
