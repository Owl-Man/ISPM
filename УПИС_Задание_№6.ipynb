## 1. Функциональные требования

### ФТ-001: Запуск новой партии
- **Описание:** Система должна позволять игроку начать новую игру, выбрав фракцию и тип противника (ИИ или человек).
- **Входные данные:** Выбор фракции, режим игры
- **Выходные данные:** Инициализированное игровое поле, стартовые карты
- **Приоритет:** Критический
- **Критерий проверки:** Игра запускается с корректными параметрами и начальным состоянием

### ФТ-002: Ход игрока
- **Описание:** Игрок должен иметь возможность совершать ход: бросить кубики, выбрать карту, построить здание.
- **Входные данные:** Команда игрока (действие), состояние руки
- **Выходные данные:** Изменение состояния игры, обновление UI
- **Приоритет:** Критический
- **Критерий проверки:** Ход выполняется корректно, изменения отражаются в интерфейсе

### ФТ-003: Ход ИИ-оппонента
- **Описание:** ИИ должен принимать решения на основе текущего состояния игры и стратегии.
- **Входные данные:** Состояние поля, карты, ресурсы
- **Выходные данные:** Действие ИИ, изменение состояния игры
- **Приоритет:** Критический
- **Критерий проверки:** ИИ совершает валидные действия, не нарушая правила

### ФТ-004: Управление картами
- **Описание:** Игрок должен иметь возможность просматривать, использовать, продавать и заменять карты.
- **Входные данные:** Выбор карты, действие
- **Выходные данные:** Обновлённая рука, ресурсы
- **Приоритет:** Высокий
- **Критерий проверки:** Карты корректно отображаются и обрабатываются

### ФТ-005: Строительство зданий
- **Описание:** Игрок может строить здания на доступных тайлах, если соблюдены условия.
- **Входные данные:** Выбор карты, тайла
- **Выходные данные:** Построенное здание, изменение ресурсов
- **Приоритет:** Высокий
- **Критерий проверки:** Здание появляется на поле, ресурсы списываются

### ФТ-006: Завершение партии
- **Описание:** Игра должна завершаться при выполнении условий победы или поражения.
- **Входные данные:** Состояние игры
- **Выходные данные:** Экран завершения, статистика
- **Приоритет:** Критический
- **Критерий проверки:** Игра корректно завершает партию и отображает результаты

### ФТ-007: Настройки игры
- **Описание:** Игрок может изменить параметры игры: звук, язык, параметры экрана.
- **Входные данные:** Выбор параметров
- **Выходные данные:** Обновлённые настройки
- **Приоритет:** Средний
- **Критерий проверки:** Настройки применяются и сохраняются

### ФТ-008: Отображение ресурсов
- **Описание:** Система должна отображать текущее количество ресурсов игрока (золото, очки).
- **Входные данные:** Состояние партии
- **Выходные данные:** Обновлённый интерфейс ресурсов
- **Приоритет:** Высокий
- **Критерий проверки:** Значения ресурсов обновляются после каждого действия

### ФТ-009: Подсказки и обучение
- **Описание:** Система должна предоставлять базовое обучение для новых игроков.
- **Входные данные:** Выбор режима обучения
- **Выходные данные:** Игровая партия с ИИ-оппонентом с пошаговым обучением основ игры
- **Приоритет:** Средний
- **Критерий проверки:** Обучение проходится полностью

### ФТ-010: Логирование ходов
- **Описание:** Система должна вести журнал действий игроков и ИИ.
- **Входные данные:** Совершённые действия
- **Выходные данные:** Записи в журнале
- **Приоритет:** Средний
- **Критерий проверки:** Все ходы фиксируются и доступны для просмотра

### ФТ-011: Система очередности ходов
- **Описание:** Система должна обеспечивать строгую очередность ходов между игроком и ИИ.
- **Входные данные:** Состояние партии
- **Выходные данные:** Передача хода следующему участнику
- **Приоритет:** Критический
- **Критерий проверки:** Ходы всегда выполняются в правильной последовательности
### ФТ-012: Система победных условий
- **Описание:** Система должна проверять выполнение условий победы (например, набор очков, контроль территории).
- **Входные данные:** Состояние партии
- **Выходные данные:** Сигнал о завершении игры
- **Приоритет:** Критический
- **Критерий проверки:** Победа фиксируется при достижении условий

### ФТ-013: Визуальные эффекты
- **Описание:** Система должна отображать анимации при ключевых действиях (розыгрыш карты, строительство).
- **Входные данные:** Игровое событие
- **Выходные данные:** Анимация
- **Приоритет:** Средний
- **Критерий проверки:** Анимации отображаются без ошибок

### ФТ-014: Звуковое сопровождение
- **Описание:** Система должна воспроизводить звуковые эффекты и музыку в зависимости от событий.
- **Входные данные:** Игровое событие
- **Выходные данные:** Звуковой эффект или музыка
- **Приоритет:** Средний
- **Критерий проверки:** Звуки соответствуют событиям и регулируются настройками

### ФТ-015: Экран статистики после партии
- **Описание:** После завершения игры система должна отображать информацию о сыгранной партии (количество построенных этажей, отсортированный по убыванию таблица игроков).
- **Входные данные:** Итоговое состояние партии
- **Выходные данные:** Экран статистики
- **Приоритет:** Высокий
- **Критерий проверки:** Статистика отображается корректно и соответствует данным партии


## 2. Нефункциональные требования

### НФТ-П-001: Использование ресурсов
- **Описание:** Игра должна стабильно работать на ПК с 6 ГБ ОЗУ и видеокартой уровня Intel HD Graphics 4000 / Radeon R7 без фризов.
- **Категория:** Производительность
- **Критерий проверки:** FPS не ниже 60

### НФТ-Н-001: Доступность
- **Описание:** Игра должна быть доступна для запуска и прохождения без сбоев
- **Категория:** Надёжность
- **Критерий проверки:** Лог ошибок показывает не более 1 критического сбоя на 100 запусков

### НФТ-Ю-001: Интуитивность интерфейса
- **Описание:** Игрок должен понимать базовые действия без обучения — начать игру, сделать ход, выбрать карту.
- **Категория:** Удобство использования
- **Критерий проверки:** 90% тестировщиков успешно совершают базовый ход без инструкций

### НФТ-С-002: Масштабируемость ИИ
- **Описание:** Сложность ИИ должна настраиваться от простого до продвинутого уровня.
- **Категория:** Масштабируемость
- **Критерий проверки:** Настройки сложности влияют на поведение ИИ в тестах

## 1. Пользовательские роли

### 1.1. Игрок
**Описание:** Конечный пользователь, запускающий игру и взаимодействующий с игровым процессом.  
**Основные задачи:**
- Запуск новой игровой партии
- Совершение игровых ходов (выбор карт, бросок кубиков, строительство зданий)
- Управление этажами главной базы

### 1.2. Разработчик
**Описание:** Член команды, выполняющий задачи по разработке
**Основные задачи:**
- Добавление и редактирование карт и игровых механик
- Изменения игрового баланса
- Отладка и исправление ошибок
- Подготовка билдов для публикации

## 2. Варианты использования (Use Cases)

### 2.1. Игрок

#### UC-001: Запуск новой партии
- **Актор:** Игрок  
- **Предусловие:** Игра установлена и запущена  
- **Сценарий:**
  1. Игрок выбирает режим игры (против ИИ или другого игрока).
  2. Игрок выбирает фракцию.
  3. Система инициализирует игровое поле и раздаёт стартовые карты.
- **Результат:** Начинается новая партия.

#### UC-002: Совершение хода
- **Актор:** Игрок  
- **Предусловие:** Игра запущена, идёт партия  
- **Сценарий:**
  1. Игрок бросает кубики.
  2. Игрок выбирает карту из руки.
  3. Игрок применяет карту или строит здание.
  4. Система обновляет состояние поля и ресурсов.
- **Результат:** Ход игрока завершён, управление передано ИИ.

#### UC-003: Настройка игры
- **Актор:** Игрок  
- **Предусловие:** Игра запущена, доступно меню настроек  
- **Сценарий:**
  1. Игрок открывает меню настроек.
  2. Изменяет параметры (звук, язык, сложность ИИ).
  3. Сохраняет изменения.
- **Результат:** Настройки применены.

#### UC-004: Завершение партии
- **Актор:** Игрок  
- **Предусловие:** Выполнены условия победы или поражения  
- **Сценарий:**
  1. Система фиксирует выполнение условий окончания.
  2. Отображает экран результатов (победа/поражение, статистика).
- **Результат:** Партия завершена.

---

### 2.2. Разработчик

#### UC-101: Добавление новой карты
- **Актор:** Разработчик  
- **Предусловие:** Доступ к редактору карт  
- **Сценарий:**
  1. Разработчик открывает модуль редактирования карт.
  2. Вводит параметры карты (название, эффект, стоимость).
  3. Сохраняет изменения.
- **Результат:** Новая карта доступна в игре.

#### UC-102: Настройка ИИ
- **Актор:** Разработчик  
- **Предусловие:** Доступ к параметрам ИИ  
- **Сценарий:**
  1. Разработчик открывает модуль конфигурации ИИ.
  2. Настраивает стратегии поведения (агрессивная, защитная и т.д.).
  3. Сохраняет изменения.
- **Результат:** ИИ использует обновлённые стратегии.

#### UC-103: Тестирование баланса
- **Актор:** Разработчик  
- **Предусловие:** Игра собрана в тестовом режиме  
- **Сценарий:**
  1. Разработчик запускает серию автоматических или ручных партий.
  2. Анализирует результаты (частота побед, эффективность карт).
  3. Вносит корректировки в параметры карт или ИИ.
- **Результат:** Баланс игры улучшен.
